# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.


Matplotlib is a Python data visualization library used for creating static, animated, and interactive visualizations in Python. It provides a comprehensive collection of functions and tools for creating a variety of charts and graphs. Matplotlib is widely used in the scientific community, data analysis, and machine learning fields.

Some reasons why Matplotlib is used are:

It is easy to use and provides high-quality visualizations.

It supports a wide range of plot types and customization options.

It integrates well with other Python libraries such as NumPy and Pandas.

It has a large and active user community that provides support and resources.

Five plots that can be plotted using the Pyplot module of Matplotlib are:

Line plot: A line plot is used to display data as a series of points connected by straight lines. It is often used to visualize trends over time.

Scatter plot: A scatter plot is used to display the relationship between two variables. It is often used to visualize the correlation between two variables.

Bar plot: A bar plot is used to compare the frequency, count, or proportion of categorical data. It is often used to visualize the distribution of data across categories.

Histogram: A histogram is used to display the distribution of a continuous variable. It is often used to visualize the frequency distribution of data.

Box plot: A box plot is used to display the distribution of a continuous variable. It is often used to visualize the spread of data and to identify outliers.

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# import numpy as np
# np.random.seed(3)
# x = 3 + np.random.normal(0, 2, 50)
# y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 500)
y = 3 + np.random.normal(0, 2, len(x))

# Create the scatter plot
plt.scatter(x, y)

# Add a title, x-label, and y-label
plt.title('Scatter Plot of X and Y')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.show()
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# Use the following data:
# import numpy as np
# For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot Line 1 on the first subplot
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Line 1')

# Plot Line 2 on the second subplot
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Line 2')

# Plot Line 3 on the third subplot
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Line 3')

# Plot Line 4 on the fourth subplot
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Line 4')

# Add a title for the entire figure
fig.suptitle('Four Line Plots using Subplot()')

# Display the plots
plt.show()
```

A bar plot is a graphical representation of categorical data in which rectangular bars are used to represent the size or frequency of each category. The height or length of each bar represents the value of the category, and the bars are usually spaced apart to distinguish between the different categories. Bar plots can be used to compare the size or frequency of different categories, to show trends over time, or to display the distribution of data.

Bar plots are used to:

Display and compare the size or frequency of different categories.
Visualize trends over time or across different groups.
Show the distribution of data.
To plot a vertical bar plot and a horizontal bar plot using the given data, we first need to import the Matplotlib library and the NumPy library, as follows:

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Vertical Bar Plot of Company Profits')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company, profit)
plt.ylabel('Company')
plt.xlabel('Profit (in millions)')
plt.title('Horizontal Bar Plot of Company Profits')
plt.show()
```

A box plot is a graphical representation of the distribution of a continuous variable through its quartiles. It is also known as a box-and-whisker plot. A box plot displays the median, quartiles, and outliers of the data distribution in a compact way. The central box represents the middle 50% of the data, and the vertical line inside the box represents the median value. The whiskers extend from the box to the minimum and maximum values that are not outliers, and any points outside of the whiskers are considered outliers.

Box plots are used to:

Visualize the distribution of a continuous variable.
Identify outliers and extreme values.
Compare the distribution of a variable across different groups or categories.
To plot a box plot using the given data, we first need to import the Matplotlib library and the NumPy library, as follows:

```python
import matplotlib.pyplot as plt
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1, box2])
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.ylabel('Value')
plt.title('Box plot of Box 1 and Box 2')
plt.show()
```
